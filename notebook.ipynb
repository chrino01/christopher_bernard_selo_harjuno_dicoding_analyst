## Explanatory Analysis & Visualization

```python
order_df.columns
```

```python
orders_customer_df = pd.merge(
    left=customer_df,
    right=order_df,
    how="right",
    left_on='customer_id',
    right_on='customer_id'

)
orders_customer_df.head()
```

```python
orders_customer_df.columns
```

```python
orders_customer_df.order_status.value_counts()
```

##### 

```python
orders_customer_df.loc[orders_customer_df['order_status'] == 'delivered'].head()
```

```python
orders_customer_df['order_purchase_timestamp'] = pd.to_datetime(orders_customer_df['order_purchase_timestamp'])
orders_customer_df['order_delivered_customer_date'] = pd.to_datetime(orders_customer_df['order_delivered_customer_date'])
orders_customer_df['order_estimated_delivery_date'] = pd.to_datetime(orders_customer_df['order_estimated_delivery_date'])

# Hitung waktu pengiriman aktual (dalam hari)
orders_customer_df['actual_delivery_time'] = (orders_customer_df['order_delivered_customer_date'] - orders_customer_df['order_purchase_timestamp']).dt.days

# Hitung waktu pengiriman yang diestimasikan (dalam hari)
orders_customer_df['estimated_delivery_time'] = (orders_customer_df['order_estimated_delivery_date'] - orders_customer_df['order_purchase_timestamp']).dt.days
```

```python
orders_customer_df.head()
```

```python
plt.figure(figsize=(10,6))

# Plot histogram distribusi waktu pengiriman aktual
sns.histplot(orders_customer_df['actual_delivery_time'], bins=30, kde=True)
plt.title('Distribusi Waktu Pengiriman Aktual (Hari)')
plt.xlabel('Waktu Pengiriman (Hari)')
plt.ylabel('Jumlah Pesanan')
plt.show()
```

#### APA PENGARUH LOKASI DARI CUSTOMER TERHADAP PENGIRIMAN DAN HARGA ONGKIR (FREIGHT VALUE)

```python
orders_customer_df.columns
```

```python
order_customer_bycity= orders_customer_df[['customer_id','customer_zip_code_prefix','customer_city','customer_state','order_id','order_status','order_purchase_timestamp','actual_delivery_time','estimated_delivery_time']]
order_customer_bycity.head()
```

```python
order_customer_bycity.customer_city.nunique()
```

```python
order_customerTimebyCity =order_customer_bycity.groupby('customer_city').agg({
    'actual_delivery_time': 'mean', 
    'estimated_delivery_time': 'mean'
    
})

byactual = order_customerTimebyCity.sort_values(by=['actual_delivery_time'], ascending=False).head(5)
byestimated = order_customerTimebyCity.sort_values(by=['estimated_delivery_time'], ascending=False).head(5)
print(byactual)
print(byestimated)
```

```python
# Menggabungkan kedua DataFrame
byactual['type'] = 'Actual'
byestimated['type'] = 'Estimated'

# Menggabungkan kedua df
df_combined = pd.concat([
    byactual.rename(columns={'actual_delivery_time': 'delivery_time'}),
    byestimated.rename(columns={'estimated_delivery_time': 'delivery_time'})
])

# Membuat bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=df_combined, x='delivery_time', y='customer_city', hue='type',orient='h')

# Mengatur label dan judul
plt.xlabel('Delivery Time (Days)')
plt.ylabel('Customer City')
plt.title('Comparison of Actual and Estimated Delivery Time for Top 5 Cities')
# plt.xticks(rotation=45)
# Menampilkan plot
plt.tight_layout()
plt.show()
```

Dari visualisasi data diatas Lokasi dari customer memiliki pengaruh yang lumayan kuat terhadap delivery time. Karena actual dari Delivery Time dari 5 kota diatas sangat lama dan berbulan bulan dari estimasi yang ada pada data. 

Untuk mengatasi ini bisa diperbaiki model algoritma ataupun aplikasi dari e commerce tersebut agar estimasi bisa lebih akurat.

## Keuntungan Top Kategori produk  ?

```python
product_df.head()
```

```python
top_five_category = ordergroup.sort_values(by='count',ascending=False).head(5)
top_five_category['total_price']= top_five_category['count'] * top_five_category['price'] 
top_five_category['total_price'] = top_five_category['total_price'].apply(lambda x: f"${x:,.1f}")
top_five_category = top_five_category.reset_index()
top_five_category
```

```python
top_five_melt = top_five_category.melt(id_vars='product_category_name',
                                    value_vars=['count','price','freight_value'],
                                    var_name= 'Type',
                                    value_name='Value')
```

```python
plt.figure(figsize=(10, 8))

# Bar plot untuk 'count' pada sumbu utama (y-axis kiri)
sns.barplot(
    data=top_five_melt[top_five_melt['Type'] == 'count'],
    x='product_category_name', y='Value', hue='Type', palette=['#1f77b4']
)

# Bar plot untuk 'price' dan 'freight_value' pada sumbu sekunder (y-axis kanan)
ax2 = plt.twinx()
sns.barplot(
    data=top_five_melt[top_five_melt['Type'].isin(['price', 'freight_value'])],
    x='product_category_name', y='Value', hue='Type', palette=['#ff7f0e', '#2ca02c'], ax=ax2
)

# Pengaturan label dan judul
plt.title("Analisa Transaksi")
plt.xlabel('Kategori Produk')
plt.ylabel('Count')  # y-axis kiri
ax2.set_ylabel('Price / Freight Value')  # y-axis kanan

# Gabungkan legenda agar tidak tumpang tindih
h, l = plt.gca().get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
plt.legend(h + h2, l + l2, loc='upper right')

plt.show()
```

```python
desired_columns = [
    'order_id', 
    'product_id', 
    'product_category_name_english',  
    'order_item_id',
    'order_status', 
    'order_purchase_timestamp', 
    'order_approved_at', 
    'order_delivered_carrier_date', 
    'order_delivered_customer_date', 
    'order_estimated_delivery_date', 
    'seller_id', 
    'shipping_limit_date', 
    'price', 
    'freight_value'
]

order_timestamp_df = order_timestamp_df[desired_columns]

order_timestamp_df.columns
```

```python
order_timestamp_df.info()
```

```python
date_columns = [
    'order_purchase_timestamp', 
    'order_approved_at', 
    'order_delivered_carrier_date', 
    'order_delivered_customer_date', 
    'order_estimated_delivery_date', 
    'shipping_limit_date'
]

for col in date_columns:
    # Mengubah kolom menjadi tipe datetime dan menghapus informasi jam
    order_timestamp_df[col] = pd.to_datetime(order_timestamp_df[col])
    order_timestamp_df[col] = order_timestamp_df[col].dt.date
    # # Memecah ke kolom tahun, bulan, hari
    # order_timestamp_df[f'{col}_year'] = pd.to_datetime(order_timestamp_df[col]).dt.year
    # order_timestamp_df[f'{col}_month'] = pd.to_datetime(order_timestamp_df[col]).dt.month
    # order_timestamp_df[f'{col}_day'] = pd.to_datetime(order_timestamp_df[col]).dt.day
```
